# Removing outliers

```python
def outlier (train_data, col):
    Q1 = train_data[col].quantile(0.25)
    Q3 = train_data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    
    #ls = df.index [ (train_data[ft] < lower_bound) | (train_data[ft] > upper_bound)];   
    ls = train_data[~((train_data[col] < (Q1 - 1.5*IQR)) | (data[col] > (Q3 + 1.5 *IQR)))]
    return ls
```

```python
index_list = []
for feature in [train_data]:
    index_list.extend(outlier(train_data,feature))
    index_list
```

```python
#define a function clled remove which returns a cleaned dataframe without outliers
def remove (train_data, ls):
    ls = sorted(set(ls))
    train_data = train_data.drop(ls)
    return train_data
```

```python
#output after removing outliers
df_cleaned = remove(train_data, index_list)
df_cleaned.shape
```
